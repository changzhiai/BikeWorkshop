## Assignment on kinetic Monte Carlo 

The Python script below implements the First Reaction Model for a kinetic Monte Carlo 
simulations of adsorption/desorption equilibrium for a lattice. The user should decide on 
size of lattice (n) and the rate constants for adsorption (ra) and desorption (rd).

Run the cell below and: 
1. Investigate how the coverage depends on the size of the lattice. Put the adsorption 
and desorption rate constants equal and change the number of lattice sites. 
2. Investigate for a give lattice how the equilibrium coverage depends on the ratio of 
the adsorption/desorption rate constants. 
3. Compare the results of the simulation with the analytical solution within the mean 
field approach: 



```python
import random 
import numpy as np

# initialize random number generator 
random.seed() 

# set parameters 
tmax = 5.0 # maximal simulation time 
n = 100 # number of sites 
ra = 1.0 # rate of adsorption 
rd = 1.0 # rate of desorption 
rate_list = [ra, rd] # list of adsorption/desorption process 


# initialize arrays/time 
t = 0 # time to 0 
told = t 
idum = 1 
tplist = [ 0. for ii in range(n)] # List of process time per site 
occ = [0 for ii in range(n)] # List of array occupation randomly 1 = occupied, 0 = empty 
no = occ.count(1) # Number of occupied sites 
nu = occ.count(0) # Number of unoccupied sites. 

print('Initial number of occupied/empty sites:', no, nu) 

# make list of process times 
for i in range(n): 
    tplist[i] = t - np.log(random.uniform(0,1))/rate_list[occ[i]] 

times = []
coverages = []
while t < tmax: 
    prind = sorted(range(n), key=tplist.__getitem__)[0] 
    prtype = occ[prind] 
    if prtype == 1: 
        # desorption 
        t = tplist[prind] 
        occ[prind] = 0 
        tplist[prind] = t - np.log(random.uniform(0,1))/rate_list[occ[prind]] 
        no = no - 1 
        nu = nu + 1 
    elif prtype == 0: 
        # adsorption 
        t = tplist[prind] 
        occ[prind] = 1 
        tplist[prind] = t - np.log(random.uniform(0,1))/rate_list[occ[prind]] 
        no = no + 1 
        nu = nu - 1 
    else: 
        print ('Site type does not exist') 
        break 
    
    theta = float(no) / float(n)
    times.append(t)
    coverages.append(theta)
    told = t 
    idum = idum + 1
```

```python
# Plot the result of the simulation
%matplotlib inline
import matplotlib.pyplot as plt

# Plot
fig, ax = plt.subplots()
ax.plot(times, coverages, label='kMC')
# Put labels on axes
ax.set_xlabel('Time')
ax.set_ylabel('Coverage')

# Try to include the analytical solution and compare
# # Analytical expression
# analyt_times = np.linspace(0, 5, 1000)
# analyt_coverage =
# ax.plot(analyt_times, analyt_coverage, 'g', label='Mean field model')

ax.legend()
print()
```
